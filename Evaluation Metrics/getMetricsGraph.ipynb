```python
import math
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

train_loss = []
dev_loss = []
EM = []
F1 = []
i = 1
f = open("AnswerGenMetrics.txt",'r')
for line in f:
    if i == 1:
        value = line[12:-1]
        train_loss.append(float(value))
        i += 1
        continue
    if i == 2:
        value = line[10:-1]
        dev_loss.append(float(value))
        i += 1
        continue
    if i == 3:
        value = line[13:-1]
        EM.append(float(value))
        i += 1
        continue
    if i == 4:
        value = line[4:-1]
        F1.append(float(value))
        i = 1
        continue
        
#print(train_loss)
#print(dev_loss)
#print(EM)
#print(F1)

t = [x for x in range(len(train_loss))]

plt.xlabel('Number of Epoch', fontsize=10)
plt.ylabel('Loss', fontsize=10)
blue_patch = mpatches.Patch(color='blue', label='Train Loss')
red_patch = mpatches.Patch(color='red', label='Validation Loss')
plt.legend(handles=[blue_patch,red_patch])
plt.plot(t, train_loss, 'b') # plotting t, a separately 
plt.plot(t, dev_loss, 'r') # plotting t, b separately 
plt.show()
    
    
    
```

```python
import matplotlib.patches as mpatches
train_loss_old = []
dev_loss_old = []
F1_old = []
EM_old = []
fp = open("oldMetrics.txt", 'r')

for line in fp:
    if line[0:11] == "Train_loss:":
        train_loss_old.append(float(line[12:-1]))
    elif line[0:9] == "Dev_loss:":
        dev_loss_old.append(float(line[10:-1]))
    elif line[0:16] == "Dev_Exact_match:":
        EM_old.append(float(line[17:-1]))
    elif line[0:13] == "Dev_F1_score:":
        F1_old.append(float(line[14:-1]))

count = 1
EM_avg = 0
avg_EM = []
for vals in EM_old:
    if count == 7:
        avg_EM.append(EM_avg*100/7.0)
        count = 1
        EM_avg = 0
    else:
        count += 1
        EM_avg += vals
print(len(avg_EM))        
t_old = [x for x in range(len(avg_EM))]
t = [x for x in range(len(EM))]

plt.xlabel('Number of Epoch', fontsize=10)
plt.ylabel('EM', fontsize=10)
blue_patch = mpatches.Patch(color='blue', label='EM New')
red_patch = mpatches.Patch(color='red', label='EM Old')
plt.legend(handles=[blue_patch,red_patch])
plt.plot(t_old, avg_EM, 'r') # plotting t, a separately 
plt.plot(t, EM, 'b') # plotting t, b separately 
       
plt.show()
```
